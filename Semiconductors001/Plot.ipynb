```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.arange(1.e-3, 4., 1.e-3)
```

```python
ys = xs * (np.log(np.exp(1. / xs) + 1) - 1. / xs)
```

```python
ym = np.log(2) * xs - 0.5
```

```python
plt.figure(figsize = (8.89, 5), dpi = 120)
plt.plot(xs, ys, 'k-', label = 'Exact')
plt.plot(xs, ym, 'b--', label = 'Approx. at $t \gg 1$')
plt.legend()
plt.xlim(0., 4.)
plt.ylim(-0.1, 2.5)
plt.xlabel(u'$t = T/(\epsilon(0) - F)$')
plt.ylabel(u'$\langle n \\rangle / G (\epsilon(0) - F)$')
plt.grid()
plt.savefig('n_from_t.pdf', bbox_inches = 'tight')
```

```python
yd = np.diff(ys) / np.diff(xs)
xd = (np.array(xs)[:-1] + np.array(xs)[1:]) / 2.
```
